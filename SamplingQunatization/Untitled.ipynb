# Sampling : spatial pos

$$
f(x,y)
$$
x, y = position (spatial intervals)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage.io import imshow, imread
```

```python
def circle_image(x, y):
    X, Y = np.meshgrid(x, y)
    return X**2 + Y**2
```

```python
factors = 2 ** np.arange(1, 6)
factors
```

```python
fig, ax = plt.subplots(1, len(factors),
                      figsize=(15, 4))
for i, N in enumerate(factors):
    image = circle_image(
        np.linspace(-1, 1, N),
        np.linspace(-1, 1, N)
    )
    ax[i].imshow(image, cmap='gray')
    ax[i].set_title("$N = {}$".format(N))
```

```python
from skimage.transform import downscale_local_mean

me = imread('me.jpg')
factors = 3 ** np.arange(1, 5)

fig, axis = plt.subplots(1, len(factors),
                      figsize=(20, 6))
for factor, ax in zip(factors, axis):
    image = downscale_local_mean(
        me,
        factors=(factor, factor, 1)
    ).astype(int)
    ax.imshow(image)
    ax.set_title("$N = {}$".format(image.shape[0]))
```
